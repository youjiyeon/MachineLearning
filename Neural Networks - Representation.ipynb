### Neural networks - notation

"activation" of unit i in layer j

i는 동일 layer 안에서의 순번이고 j는 layer 번호입니다.

또한 a는 layer j의 unit i의 activation(출력 값)을 의미합니다. 

a는 g로 표시되는 logistic activation function를 통해 나온 값을 의미합니다.//뒤에 나오는 내용(강조) 

수식에서와 같이 각 입력에 대한 Θ가 곱한한 값을 logistic activation function에 넣었을 때 나오는 결과가 바로 activation이고 a로 나타냅니다.



Θ(j)=matrix of weights controlling function mapping from layer j to layer j+1

Θ는 layer j에서 j+1로 입력이 들어갈때 가중치를 정하는 함수입니다.

Θ의 윗첨자는 j에서 j+1 Layer로 넘어갈때 발생한다는 의미입니다.

즉  θ(1)은 layer 1 에서 layer 2로 넘어가면서 발생했다는 뜻입니다. 

아랫첨자는 각자 layer j+1에서의 순번(i)와  layer j+1의 순번(i)를 의미합니다. 

즉,  θ12는 a1과 x2간의 가중치라는 의미입니다.

j는 도착한 unit의 번호

i는 보낸 unit의 번호 입니다.

결국 Layer j에 속한 unit의 갯수를 sj. 

Layer j+1에 속한 unit의 갯수를 sj+1라고 표현한다면  θ(j)의 갯수는  sj+1 x (sj +1) 이 됩니다.
